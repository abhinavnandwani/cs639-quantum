### Part A Solution

<span style="color: red;"> The 
No-Cloning Theorem states that an arbitrary unknown quantum state cannot be copied perfectly. The standard proof assumes an initial auxiliary state $|0\rangle_B$, and we analyze whether replacing this with an arbitrary state $|\phi\rangle_B$ affects the validity of the theorem.  
Proof Attempt  
A cloning operation, if possible, would be represented by a unitary transformation $U_{\text{clone}}$ such that for any two arbitrary states $|\psi\rangle_A$ and $|\phi\rangle_B$:  
$U_{\text{clone}} (|\psi\rangle_A \otimes |\phi\rangle_B) = |\psi\rangle_A \otimes |\psi\rangle_B$.  
For two different states $|\psi\rangle$ and $|\psi'\rangle$, we require:  
$U_{\text{clone}} (|\psi\rangle_A \otimes |\phi\rangle_B) = |\psi\rangle_A \otimes |\psi\rangle_B$,
$U_{\text{clone}} (|\psi'\rangle_A \otimes |\phi\rangle_B) = |\psi'\rangle_A \otimes |\psi'\rangle_B$.  
Taking the inner product of these two equations:  
$\langle \psi | \psi' \rangle \langle \psi | \psi' \rangle = \langle \psi | \psi' \rangle$.  
Since $|\psi\rangle$ and $|\psi'\rangle$ are arbitrary, this equation holds only if $\langle \psi | \psi' \rangle = 1$ or $\langle \psi | \psi' \rangle = 0$, meaning only orthogonal states could be cloned.  
Thus, even if we replace $|0\rangle_B$ with an arbitrary state $|\phi\rangle_B$, the proof structure remains the same, and the no-cloning theorem still holds.

The no-cloning theorem prevents the perfect copying of an arbitrary unknown quantum state. However, if the state to be cloned is known and belongs to an orthonormal basis (such as ${|0\rangle, |1\rangle}$), classical copying is possible using a CNOT gate.  
Cloning Circuit for $|1\rangle$  
A simple quantum circuit that clones $|1\rangle$ when the second qubit is initialized to $|0\rangle$ consists of:  
A CNOT gate with the first qubit as the control and the second qubit as the target.
The circuit implementation of $U_{\text{clone}}$ is:  
Input: $|1\rangle_A \otimes |0\rangle_B$.
Apply CNOT gate: The target qubit flips if the control qubit is $|1\rangle$.  
Output: $|1\rangle_A \otimes |1\rangle_B$.  
This works because the computational basis states ${|0\rangle, |1\rangle}$ are orthogonal and can be copied using classical logic.
However, for a general quantum state $|\psi\rangle = \alpha |0\rangle + \beta |1\rangle$, the CNOT gate does not preserve superposition, reinforcing the no-cloning theorem.

### Part D Solution
What is the property of the operator $U_{clone}$ that restricts it from cloning arbitrary states? Either provide the name of the property or write out the definition of the property as equations involving $U_{clone}$.

<span style="color: red;"> The 
property that restricts $U_{clone}$ from cloning arbitrary states is **linearity** of quantum operations.
A unitary operator $U_{clone}$ must satisfy linearity, meaning that for any two quantum states $|\psi\rangle$ and $|\phi\rangle$ and any complex coefficients $a, b$:  
$U_{clone} (a |\psi\rangle + b |\phi\rangle) = a U_{clone} |\psi\rangle + b U_{clone} |\phi\rangle$.  
If perfect cloning were possible, then for any arbitrary states $|\psi\rangle$ and $|\phi\rangle$, we would require:  
$U_{clone} |\psi\rangle = |\psi\rangle \otimes |\psi\rangle, \quad U_{clone} |\phi\rangle = |\phi\rangle \otimes |\phi\rangle$.  
Applying linearity to a superposition state $(a|\psi\rangle + b|\phi\rangle)$, we get:  
$U_{clone} (a |\psi\rangle + b |\phi\rangle) = a U_{clone} |\psi\rangle + b U_{clone} |\phi\rangle = a |\psi\rangle \otimes |\psi\rangle + b |\phi\rangle \otimes |\phi\rangle$.  
However, if cloning were truly possible, we would also expect:  
$U_{clone} (a |\psi\rangle + b |\phi\rangle) = (a |\psi\rangle + b |\phi\rangle) \otimes (a |\psi\rangle + b |\phi\rangle)$.  
These two results are not equal in general, violating linearity. Since quantum mechanics requires all valid transformations to be linear, this contradiction proves that a universal $U_{clone}$ cannot exist, reinforcing the no-cloning theorem.
</span>
