## Q2: Entanglement

### Part A Solution
Consider the quantum state $$\frac{1}{2}(|00 \rangle  - i|01⟩ + i|10⟩ + |11⟩)$$

Is this state entangled or separable? If entangled, explain how you know using the probabilities of measurement outcomes. If separable, provide a decomposition into a product of component spaces.

<span style="color: red;"> We 
need to determine whether the given two-qubit state $\frac{1}{2}(|00\rangle - i|01\rangle + i|10\rangle + |11\rangle)$ is entangled or separable.  
A state is separable if it can be written as a tensor product of two single-qubit states:
$|\psi\rangle = |\psi_A\rangle \otimes |\psi_B\rangle$.
Otherwise, it is entangled.  
Step 1: Check for Factorization  
Suppose the state is separable, meaning there exist single-qubit states  
$(a_0|0\rangle + a_1|1\rangle) \otimes (b_0|0\rangle + b_1|1\rangle)$.  
Expanding,  
$(a_0b_0)|00\rangle + (a_0b_1)|01\rangle + (a_1b_0)|10\rangle + (a_1b_1)|11\rangle$.  
Comparing with the given state:  
$a_0b_0 = \frac{1}{2}$,  
$a_0b_1 = -\frac{i}{2}$,  
$a_1b_0 = \frac{i}{2}$,  
$a_1b_1 = \frac{1}{2}$.  
From these equations, solving for $a_0$ and $a_1$, and comparing consistency, we find that no valid solutions exist where  
$a_0, a_1, b_0, b_1$ satisfy all equations simultaneously.  
Step 2: Compute Probabilities for Measurement Outcomes  
The probability of measuring a given basis state is the squared magnitude of its coefficient. The probabilities are:  
$P_{00} = \left| \frac{1}{2} \right|^2 = \frac{1}{4}$,  
$P_{01} = \left| -\frac{i}{2} \right|^2 = \frac{1}{4}$,  
$P_{10} = \left| \frac{i}{2} \right|^2 = \frac{1}{4}$,  
$P_{11} = \left| \frac{1}{2} \right|^2 = \frac{1}{4}$.    
Since these probabilities do not allow the state to be written as a product of independent single-qubit probabilities, the state is entangled.  
</span>

initial state is $|000\rangle$.
Step 1: Apply Hadamard Gate to the First Qubit  
The Hadamard gate transforms $|0\rangle$ into a superposition:
$H|0\rangle = \frac{|0\rangle + |1\rangle}{\sqrt{2}}$.  
After this operation, the state becomes
$\frac{1}{\sqrt{2}} (|000\rangle + |100\rangle)$.  
Step 2: Apply CNOT from Qubit 1 to Qubit 2  
The CNOT flips the second qubit if the first qubit is $|1\rangle$:
$\frac{1}{\sqrt{2}} (|000\rangle + |110\rangle)$.  
Step 3: Apply CNOT from Qubit 2 to Qubit 3  
The CNOT flips the third qubit if the second qubit is $|1\rangle$:
$\frac{1}{\sqrt{2}} (|000\rangle + |111\rangle)$.  
This is the GHZ state, which is entangled because it cannot be written as a product of independent single-qubit states.  
Measurement Probabilities  
Measuring in the computational basis gives either $|000\rangle$ or $|111\rangle$ with equal probability $\frac{1}{2}$. Measuring any one qubit collapses the entire state, confirming entanglement.

### Part C Solution
Create a two qubit circuit with at least one Hadamard gate and one CNOT gate such that the qubits are NOT entangled after applying the circuit (starting from the state $|00⟩$).

<span style="color: red;"> We 
need to construct a two-qubit circuit with at least one Hadamard gate and one CNOT gate such that the qubits remain **separable** after applying the circuit.
Step 1: Apply Hadamard to the First Qubit  
The Hadamard gate transforms $|0\rangle$ into a superposition:
$H|0\rangle = \frac{|0\rangle + |1\rangle}{\sqrt{2}}$.  
After this operation, the state is $\frac{|00\rangle + |10\rangle}{\sqrt{2}}$.  
Step 2: Apply CNOT (First Qubit as Control, Second as Target)  
The CNOT flips the second qubit when the first qubit is $|1\rangle$, resulting in $\frac{|00\rangle + |11\rangle}{\sqrt{2}}$.  
Step 3: Apply Hadamard to the Second Qubit  
The Hadamard gate acts as $H|0\rangle = \frac{|0\rangle + |1\rangle}{\sqrt{2}}$ and $H|1\rangle = \frac{|0\rangle - |1\rangle}{\sqrt{2}}$.  
Applying this to the second qubit, the final state becomes $\frac{|0\rangle + |1\rangle}{\sqrt{2}} \otimes \frac{|0\rangle - |1\rangle}{\sqrt{2}}$.  
Since the final state is a tensor product of two independent qubit states, the qubits are not entangled.
</span>
